### Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).

```python
stock_data = stock_data.set_index('stock', append= True)
```

```python
stock_data
```

### Create a dataFrame called vol, with the volume values.

```python
volume = stock_data['Volume']
```

```python
volume.head(7)
```

### Aggregate the data of volume to weekly.
Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
stock_data=stock_data.reset_index()
```

```python
stock_data
```

```python
stock_data =stock_data.set_index('Date')
```

```python
stock_data
```

```python
stock_data.groupby('stock').resample('W').sum()
```

```python
stock_data =stock_data.groupby('stock').resample('W').Volume.sum().reset_index()
```

```python
stock_data
```

```python
volume = stock_data.pivot_table(  values= 'Volume', index= 'Date' , columns= 'stock')
```

```python
volume
```

```python
volume2 = volume.resample('W').sum()
```

```python
volume2
```

### Find all the volume traded in the year of 2015

```python
volume2.reset_index(inplace=True)
```

```python
volume2 =volume2.set_index('Date').resample('Y').sum()

```
