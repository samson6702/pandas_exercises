### Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data). 

### Assign it to a variable called iris

```python
import pandas as pd
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url)

iris.head()
```

### Create columns for the dataset

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class



iris.columns =['sepal_length','sepal_width','petal_length','petal_width','class']
```

```python
iris.columns
```

```python
iris
```

###  Is there any missing value in the dataframe?

```python
iris.isna()
```

```python
iris.isnull().sum()
```

###  Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
import numpy as np
```

```python
null = np.NAN
iris.iloc[10:30,2] =null
iris.head(12)
```

### Good, now lets substitute the NaN values to 1.0

```python
iris.fillna(1.0 ,inplace=True ,axis=1)
```

```python
iris.head(12)
```

### Now let's delete the column class

```python
iris.drop(labels='class' ,axis=1)
```

###  Set the first 3 rows as NaN

```python
iris.iloc[:3] =null
```

```python
iris.head(5)
```

###  Delete the rows that have NaN

```python
iris =iris.dropna()
```

```python
iris.head(12)
```

### Reset the index so it begins with 0 again

```python
iris =iris.reset_index(drop=True)
```
